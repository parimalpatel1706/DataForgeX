#### MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## simple examples
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
### plt scatter
plt.scatter(x,y,c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
y=x*x
```

```python
##plt plot
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('straight line graph')
```

```python
plt.plot(x,y,'gs',linestyle='dashed',linewidth=2,markersize=12,mfc='r')#mfc is used for ki pura marker ko colour karne ke liye
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('y=x^2 graph')
```

```python
### creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')# o is used for scatter plot
```

```python
## some complex operations
x=np.arange(1,11)
y= 3 * x + 5
plt.title("Matplotlib demo")
plt.xlabel('x captios')
plt.ylabel('y captions')
plt.plot(x,y)
plt.show()
```

```python
### creating a sine wave form
x=np.arange(0, 4 *np.pi,0.1)
y=np.sin(x)## sin() is a built function
plt.title("Sine Wave form")
plt.plot(x,y)
plt.grid()
```

```python
## using subplot to create a sin and cos wave form
x=np.arange(0,5 *np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
## for sin wave
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title("sine wave")
## for cos wave
plt.subplot(2,1,2)
plt.plot(x,y_cos,'m*')
plt.title("cos wave")
```

#### Bar Plot

```python
x1=[2,8,10]
y1=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x1,y1)
plt.bar(x2,y2)
plt.title("bar graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

#### Histograms

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,color='red',bins=20)
plt.title("histogram")
```

#### Box Plot

```python
data= [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

#### Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
