```python
import pandas as pd
```

read_html:-ek Pandas function hai jo kisi HTML page (ya file) mein se tables ko automatically detect karke DataFrame mein convert karta hai.

```python
html=pd.read_html(r"https://en.wikipedia.org/wiki/Mobile_country_code")### basically in html file read function read table rows and table header usko consider karta hai aur pure table ko read kar deta hai
```

```python
type(html)
```

```python
## agar hame data set se specific table uthna hai toh hum "match" ka use karte hai
html=pd.read_html("https://en.wikipedia.org/wiki/Economy_of_the_United_States",match="Government debt")
```

```python
html[0]
```

```python
type(html[0])
```

```python
html=pd.read_html("https://en.wikipedia.org/wiki/Minnesota",match="Average daily maximum and minimum temperatures for selected cities in Minnesota")
```

```python
html[0]
```

```python
type(html[0])
```

```python
html[0].to_html('demo.html')## data frame ko html format mein convert karte hai
```
