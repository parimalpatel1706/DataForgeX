```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
df.head()
```

```python
## joint plot
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='reg')
```

```python
## pair plot
sns.pairplot(df,hue="species")
```

```python
##Distplot
sns.displot(df['sepal_length'])
```

```python
## countplot
sns.countplot(x='species',data=df,hue='species')
```

```python
## bar plot
sns.barplot(x='species',y='sepal_length',data=df,hue='species')
```

```python
## box plot
sns.boxplot(x='sepal_length',y='species',hue='species',data=df)
```

```python
## violin plot
sns.violinplot(x='sepal_length',y='species',data=df,palette='rainbow')
```
